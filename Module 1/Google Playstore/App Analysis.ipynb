```python
import plotly.offline as pyo
import plotly.graph_objs as go

import numpy as np
import pandas as pd

data = pd.read_csv('googleplaystore.csv')
data = data.iloc[:,1:]
```

```python
data.head(100)
```

```python
categories_names = np.unique(data.iloc[:,0:1].values)
categories_data  = data.iloc[:,0:1].values

ratings = data.iloc[:,1:2].values
ratings = np.reshape(ratings, ratings.shape[0])

type_ = data.iloc[:,5:6].values
type_ = np.reshape(type_, type_.shape[0])

content_rating = data.iloc[:,7:8].values
content_rating = np.reshape(content_rating, content_rating.shape[0])



def get_ratings(data):
    
    ref0  = ref1_ = ref1 = ref2_ = ref2 = ref3_ = ref3 = ref4_ = ref4 = ref5 = 0
    return_data = {}
    
    for i in range(len(ratings)):
        
        if ratings[i] <= 1.0: ref0 += 1
        if ratings[i] >= 1.0 and ratings[i] < 1.5 : ref1 += 1
        if ratings[i] >= 1.5 and ratings[i] < 2.0 : ref1_+= 1
        if ratings[i] >= 2.0 and ratings[i] < 2.5 : ref2 += 1
        if ratings[i] >= 2.5 and ratings[i] < 3.0 : ref2_+= 1
        if ratings[i] >= 3.0 and ratings[i] < 3.5 : ref3 += 1
        if ratings[i] >= 3.5 and ratings[i] < 4.0 : ref3_+= 1
        if ratings[i] >= 4.0 and ratings[i] < 4.5 : ref4 += 1
        if ratings[i] >= 4.5 and ratings[i] < 5.0 : ref4_+= 1
        if ratings[i] == 5.0: ref5 += 1
    print('      Ratings Analysis      ')
    print('****************************')
    print('0   - 1.0\t:',ref0 )
    print('1.0 - 1.5\t:',ref1 )
    print('1.5 - 2.0\t:',ref1_)
    print('2.0 - 2.5\t:',ref2 )
    print('2.5 - 3.0\t:',ref2_)
    print('3.0 - 3.5\t:',ref3 )
    print('3.5 - 4.0\t:',ref3_)
    print('4.0 - 4.5\t:',ref4 )
    print('4.5 - 5.0\t:',ref4_)
    print('  5.0    \t:',ref5) 
    print('****************************')
    print('')
    
    return_data['0   - 1.0'] = ref0
    return_data['1.0 - 1.5'] = ref1
    return_data['1.5 - 2.0'] = ref1_
    return_data['2.0 - 2.5'] = ref2
    return_data['2.5 - 3.0'] = ref2_
    return_data['3.0 - 3.5'] = ref3
    return_data['3.5 - 4.0'] = ref3_
    return_data['4.0 - 4.5'] = ref4
    return_data['4.5 - 5.0'] = ref4_
    return_data['  5.0    '] = ref5

    return return_data


def get_categories():
    print('    Categories Analysis    ')
    print('****************************')
    for i in range (1,len(categories_names)):
        ref = 0
        for j in range(len(categories_data)):

            if categories_data[j] == categories_names[i]:
                ref = ref + 1
            else:
                pass

        print(ref,'\t : \t',categories_names[i])
    print('****************************')
    print('')

def get_type():
    print('     App type Analysis    ')
    print('****************************')
    free = paid = 0
    for i in range(len(type_)):
        if type_[i] == 'Free': free += 1
        else: paid += 1

    print('\t Free: ',free)
    print('\t Paid: ',paid)
    print('****************************')
    print('')
    

def get_content_rating():
    print('  Content Rating Analysis    ')
    print('****************************')

    everyone = teen = everyone10 = mature = adult = 0

    for i in range(len(content_rating)):

        if content_rating[i] == 'Adults only 18+': adult   += 1
        if content_rating[i] == 'Everyone'       : everyone   += 1
        if content_rating[i] == 'Teen'           : teen       += 1
        if content_rating[i] == 'Mature 17+'     : mature   += 1
        if content_rating[i] == 'Everyone 10+'   : everyone10 += 1

    print('Everyone         : ',everyone)
    print('Everyone 10+     : ',everyone10)
    print('Teen             : ',teen)
    print('Adults only      : ',adult)
    print('Mature 17+       : ',mature)
    print('****************************')
    print('')
```

```python
data = get_ratings(ratings)
# get_categories()
# get_content_rating()
# get_type()
```

```python
keys   = list(data.keys())
values = list(data.values())

graph  = [go.Bar(x = keys, y = values)]
layout = go.Layout(title = 'Analysis')

fig = go.Figure(graph, layout)
pyo.plot(fig)
```
