```python
import os
import dlib
import cv2 as cv

import matplotlib.pyplot as plt

import numpy as np

face_detector = dlib.get_frontal_face_detector()
```

```python
imgs = os.listdir('Dataset/Ashish/')

for img in imgs:
    
    im = cv.imread("Dataset/Not Ashish/"+img)
    
    gray = cv.cvtColor(im, cv.COLOR_BGR2GRAY)
    
    faces = face_detector(gray)
    
    for face in faces:
        
        x1 = int(face.left())
        y1 = int(face.top())
        x2 = int(face.right())
        y2 = int(face.bottom())
        
        crop = im[y1:y2,x1:x2,:]
        
        
    cv.imwrite('Dataset/Cropped/'+img, crop)
```
