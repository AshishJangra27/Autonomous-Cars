```python
import os
import serial
import cv2 as cv
```

```python
s = serial.Serial("/dev/cu.usbmodem14101", baudrate = 9600)
cam = cv.VideoCapture(0)

ref = len(os.listdir('IMG'))

fd = open('data_log.txt','a+')

while True:
    
    ref += 1
    _, img = cam.read()
    
    path = "IMG/" + str(ref) + '.jpg'
    cv.imwrite(path,img)
    
    s.write('1')
    
    angle = s.readline()
    
    data = path + ',' + str(angle)
    fd.write(data)
    
    print(angle)
    
    if(ref == 1100):
        break


cam.release()
s.close()
fd.close()
```
