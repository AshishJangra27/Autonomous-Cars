```python
import cv2 as cv
from serial import Serial
from keras.models import load_model

import numpy as np
import pandas as pd

ser = Serial('/dev/cu.usbmodem14101' , baudrate = 9600, timeout = 0.0001)

model = load_model('Best1.h5')

cap = cv.VideoCapture('Simulator.mov')

while True:

    ret, frame = cap.read()
    frame = cv.resize(frame, (320*3,160*3))
    cv.imshow('Frame',frame)

    frame = cv.resize(frame, (160,320))
    frame = np.reshape(frame, (1,160,320,3))


    angle = str(int(model.predict(frame)[0][0]*600 + 45))

    ser.write(angle)

    print(angle)

    if cv.waitKey(1) & 0xFF == ord('q'):
        
        break
    
cap.release()
cv.destroyAllWindows()
```
