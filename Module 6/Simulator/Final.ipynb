```python
import cv2 as cv
import numpy as np
from serial import Serial
from keras.models import load_model

model = load_model('Best1.h5')
ser = Serial('/dev/cu.usbmodem14101', baudrate = 9600, timeout = 0.0001)
```

```python
cap = cv.VideoCapture("Simulator.mov")

while True:
    
    _, frame = cap.read()

    cv.imshow("DashCam",frame)
    
    frame = cv.resize(frame , (160,320))
    
    frame = np.reshape(frame, (1,160,320,3))
    
    angle = str(int((model.predict(frame)[0][0] + 1.)*90.)) 
    ser.write(angle)

    
    print(angle)
    
    if cv.waitKey(30) == ord('q'):
        break
```
